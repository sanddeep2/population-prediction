```python
# X=[2017,2018,2019,2020,2021]
X = a['year'].tolist()
Ygirls = a['female'].tolist()
# Ygirls = [20,25,30,20,30]
Zboys = a['male'].tolist()
# Zboys = [12,30,52,60,25]
  
X_axis = np.arange(len(X))
plt.figure(figsize=(16,8))
plt.bar(X_axis-0.2, Ygirls,0.4, label = 'Girls',color='b')
plt.bar(X_axis+0.2, Zboys,0.4, label = 'Boys',color='r')
  
plt.xticks(X_axis, X)
plt.xlabel("year")
plt.ylabel("population")
plt.title("male vs female")
plt.legend()
plt.show()
```

```python
from scipy import stats

x = a['year'].tolist()
y = a['population'].tolist()

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()

print("relationship = ",r)

x1=myfunc(2022)
print("population in 2022 = ", x1)

x2=myfunc(2023)
print("population in 2023 = ", x2)

x3=myfunc(2024)
print("population in 2024 = ", x3)

x4=myfunc(2025)
print("population in 2025 = ", x4)
```

```python
x = a['year'].tolist()
y = a['population'].tolist()

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(2017, 2025, 1500000)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

z1=mymodel(2022)
print("population in 2022 = ", z1)

z2=mymodel(2023)
print("population in 2023 = ", z2)

z3=mymodel(2024)
print("population in 2024 = ", z3)

z4=mymodel(2025)
print("population in 2025 = ", z4)
```
